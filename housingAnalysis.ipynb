# Data Analysis on the Housing Website Info

### Eats data from housing availability website -- Spits out some info?


```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv("housingTwo.csv")
data
```

```python
data.drop(["Web Link", "Single", "Double", "2", "3", "4", "5", "6", "7", "8"], axis = 1, inplace = True)
```

```python
data.columns = ["Building", "Single", "Double", "DateTime"]
data
```

```python
data.sort_values(["Building", "DateTime"], ignore_index = True, inplace = True)
data = data[:-20]
data
```

```python
data.set_index("DateTime", inplace = True)
```

```python
data = data[data.Single != "Single"]
data.Single = data.Single.astype(float)
data.Double = data.Double.astype(float)
by_building = data.groupby(["Building"]).diff(axis = 1)
```
